```python
import math
import matplotlib.pyplot as plt
import astropy.units as u
from scipy import stats
from astropy.coordinates import SkyCoord
```

Diagramas cor-magnitude dos aglomerados (antes dos ajustes)

```python
infile1 = "/content/Fis2013a.dat"
infile2 = "/content/Fis2013b.dat"

file1 = open(infile1, "r")

als_BV=[]
als_V=[]
bls_BV=[]
bls_V=[]

next(file1)             #Pula a primeira linha do arquivo (cabeçalho)
                        
for linha in file1:
  partes = linha.split()
  als_BV.append(float(partes[7].strip()))
  als_V.append(float(partes[3].strip()))

file1.close()

file2 = open(infile2, "r")

next(file2)             #Pula a primeira linha do arquivo (cabeçalho)
                        
for linha in file2:
  partes = linha.split()
  bls_BV.append(float(partes[7].strip()))
  bls_V.append(float(partes[3].strip()))

file2.close()
```

```python
#Fis02013a
plt.scatter(als_BV, als_V, s=1)
plt.gca().invert_yaxis()
plt.title("Fis02013a")
plt.ylabel("$V_{mag}$")
plt.xlabel("B-V")
plt.show()
```

```python
#Fis02013b
plt.scatter(bls_BV, bls_V, s=1)
plt.gca().invert_yaxis()
plt.title("Fis02013b")
plt.ylabel("$V_{mag}$")
plt.xlabel("B-V")
plt.show()
```

Procurando isócronas para Fis02013a

```python
infile3 = "/content/bvr_008_4750Myr.dat"

file3 = open(infile3, "r")

iso_BV=[]
iso_V=[]
                        
for linha in file3:
  partes = linha.split()
  iso_BV.append(float(partes[4].strip()))
  iso_V.append(float(partes[2].strip()))

file3.close()
```

```python
#Comparando com Fis02013a
plt.scatter(als_BV, als_V, s=1, label="Fis02013a")
plt.scatter(iso_BV, iso_V, s=1, c="red", label="Isócrona: 4.75Gyr" + "\n" + "Z=0.08")
plt.gca().invert_yaxis()
plt.ylabel("$V_{mag}$")
plt.xlabel("B-V")
plt.legend()
plt.show()
```

```python
#Ajustando
ajustea_BV=[]
ajustea_V=[]

EBV=0.25
mv_Mv=10

for i in range(len(als_BV)):
  ajustea_BV.append(als_BV[i] - EBV)
for i in range(len(als_V)):
  ajustea_V.append(als_V[i] - mv_Mv)

plt.scatter(ajustea_BV, ajustea_V, s=1, label="Fis02013a")
plt.scatter(iso_BV, iso_V, s=1, c="red", label="Isócrona: 4.75Gyr" + "\n" + "Z=0.08")
plt.gca().invert_yaxis()
plt.ylabel("$V_{mag}$")
plt.xlabel("B-V")
plt.legend()
plt.show()
```
